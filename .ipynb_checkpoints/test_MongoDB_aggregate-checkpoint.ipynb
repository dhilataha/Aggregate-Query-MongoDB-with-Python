```python
#Menampilkan Jumlah Film Setiap Category
hasil = db.Sakila_films.aggregate([{"$group":{"_id":"$Category", "Jumlah Film":{"$sum": 1}}}])
DataFrame(hasil)
```

```python
#2 - Menampilkan Jumlah Judul Film Setiap Rating
hasil = db.Sakila_films.aggregate([{"$group": {"_id": "$Rating", "Jumlah Film": {"$sum": 1}}}])
DataFrame(hasil)
```

```python
#3 - Menampilkan Jumlah Actors dalam setiap Film dengan Rating PG
hasil = db.Sakila_films.aggregate([
    {"$match": {"Rating" : "PG"}},
    {"$unwind" : "$Actors"},
    {"$group": {"_id": "$Title", "Jumlah Actor": {"$sum": 1}}}
])
DataFrame(hasil)
```

```python
#4 - Menampilkan Total Perentalan Film setiap Customer di United States
hasil = db.Sakila_customers.aggregate([
    {"$match": {"Country" : "United States"}},
    {"$unwind" : "$Rentals"},
     {
          "$project" : {
              "First Name" : 1,
              "Last Name" : 1
          }
    },
    {
        "$group" : {
            "_id" : "$First Name",
            "LastName" : {
                "$first" : "$Last Name"
            },
            "Total Rental" : {"$sum":1}
        }
    },
])
DataFrame(hasil)
```

```python
#5 - Menghitung Rata-rata Biaya yang Dikeluarkan Setiap Negara untuk Merental Film
datas = db.Sakila_customers.aggregate([
    {
        "$unwind" : "$Rentals"
    },
    {
        "$project" : {
            "_id" : 1,
            "Country" : 1,
            "Payments" : {"$sum":"$Rentals.Payments.Amount"}
        }
    },
    {
        "$group" : {
            "_id" : "$Country",
            "CustomersAveragePayments" : {"$push" : "$Payments"}
        }
    }
])

listDatas = list(datas) #Mengubah (cast) commandCursor menjadi list
newDatas = [] #Inisialisasi list baru


for i in range(len(listDatas)):
    paymentsLen = len(listDatas[i]['CustomersAveragePayments']) #Mendapatkan panjang payments
    paymentsByCountry = 0.0 #Insialisasi variabel yang akan dijumlahkan
    for j in range(paymentsLen):
        #Menjumlahkan semua payments berdasarkan negara
        paymentsByCountry = paymentsByCountry + listDatas[i]['CustomersAveragePayments'][j] 

    #Memasukkann data negara dan rata2 payments ke dalam list baru (newDatas)
    newDatas.append({
        "Country" : listDatas[i]['_id'],
        "CustomersAveragePayments" : paymentsByCountry/paymentsLen
    })
DataFrame(newDatas)
```

```python
#6 - Menampilkan Total Payment Amount Terbanyak dari Customers
datas = db.Sakila_customers.aggregate(
    [
        {
            "$unwind" : "$Rentals"
        },
        {
            "$project" : {
                "_id" : 1,
                "First Name" : 1,
                "Last Name" : 1,
                "Payments" : {"$sum":"$Rentals.Payments.Amount"}
            }
        },
        {
            "$group": {
                "_id": {
                    "First Name" : "$First Name",
                    "Last Name" : "$Last Name"
                },
                "Rentals": {"$push":"$Payments"}
            }
        },
        
    ]
)

listDatas = list(datas)
dataLen = len(listDatas)
newDatas = []
totalPayments = []
for x in range(dataLen):
    rentalLen = len(listDatas[x]['Rentals'])
    totalPayment = 0.0
    for y in range(rentalLen):
        totalPayment = totalPayment + listDatas[x]['Rentals'][y]
    newDatas.append({"id":listDatas[x]['_id'], "TotalPayment" : totalPayment})
    totalPayments.append(totalPayment)
    
totalPayments.sort(reverse=True)
sortedTotalPayments = []

for i in range(10):
    for j in range(len(totalPayments)):
        if totalPayments[i] == newDatas[j]['TotalPayment']:
            sortedTotalPayments.append(newDatas[j])
DataFrame(sortedTotalPayments)
```
