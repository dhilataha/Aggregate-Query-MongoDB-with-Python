```python
from pymongo import MongoClient
client = MongoClient()
```

```python
db = client.sakila
```

```python
#1
description = db.Sakila_films.aggregate([{"$match": { "$or" : [{"Description" : {"$regex" : "Shark"}},{"Description" : {"$regex" : "Crocodile"}}]}},{"$group" : {"_id" : "Description","Total Film" : {"$sum":1}}}])
for a in description:
    print(a)
```

```python
#2
rating = db.Sakila_films.aggregate([{"$group": {"_id": "$Rating", "Jumlah Film": {"$sum": 1}}}])
for a in rating:
    print(a)
```

```python
#3
actors = db.Sakila_films.aggregate([
    {"$match": {"Rating" : "PG"}},
    {"$unwind" : "$Actors"},
    {"$group": {"_id": "$Title", "Jumlah Actor": {"$sum": 1}}}
])
for a in actors:
    print(a)
```

```python
#4 
us = db.Sakila_customers.aggregate([
    {"$match": {"Country" : "United States"}},{"$unwind" : "$Rentals"},{"$project" : {"First Name" : 1,"Last Name" : 1}},
    {"$group" : {"_id" : "$First Name","LastName" : {"$first" : "$Last Name"},"Total Rental" : {"$sum":1}}},])
for a in us:
    print(a)
```
